```python
import pandas as pd

# --- Load datasets ---
daily_factors = pd.read_csv("merged_daily_factors_final.csv", parse_dates=["date"])
price_data = pd.read_csv("AAPL.csv", parse_dates=["Date"])

# --- Standardize and clean ---
price_data.rename(columns={"Date": "date", "Adj Close": "adj_close"}, inplace=True)
price_data['date'] = price_data['date'].dt.normalize()
daily_factors['date'] = pd.to_datetime(daily_factors['date']).dt.normalize()

# --- Subset to only needed columns ---
price_data = price_data[['date', 'adj_close']].dropna()

# --- Merge datasets ---
df_merged = pd.merge(daily_factors, price_data, on="date", how="left")

# --- Diagnostics ---
missing = df_merged['adj_close'].isna().sum()
print(f"âœ… Merged dataset shape: {df_merged.shape}")
print(f"ðŸ§ª Missing 'close' values after merge: {missing}")

# Optional: drop rows without close prices
df_merged = df_merged.dropna(subset=["adj_close"]).reset_index(drop=True)
```
